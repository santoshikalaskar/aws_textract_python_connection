```python
import boto3
```

```python
boto3.__version__
```

```python
client = boto3.client('textract')
```

```python
img = 'Data/test.jpg'

with open(img, 'rb') as raw_image:
    img_temp = raw_image.read()
    bytes_img = bytearray(img_temp)
    

response = client.detect_document_text(Document = {'Bytes': bytes_img})
```
